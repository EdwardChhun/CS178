```python
import pandas as pd
import numpy as np

seed = 1234
np.random.seed(seed)
```

```python
df = pd.read_csv("training_data.csv")

print(df.shape)
df.head()
```

```python
"""
Replacing the missing data with NaN
"""

missing_cols = ['race', 'weight', 'payer_code', 'medical_specialty', 'diag_1', 'diag_2', 'diag_3', 'gender']

column_to_missing = {}
for col in missing_cols:
    if col != 'gender':
        column_to_missing[col] = '?'
    else:
        column_to_missing[col] = 'Unknown/Invalid'

# print(column_to_missing)

for col, miss_val in column_to_missing.items():
    df[col] = df[col].replace(miss_val, np.nan)
print("Before Mode Impute:\n================")
df.isnull().sum()
```

```python
"""
Mode Imputation
"""
df_1 = df.copy()
mode_impute_col_names = ['race', 'payer_code', 'medical_specialty']

def mode_impute(df_column):
    mask = df_column.notna()
    mode = df_column[mask].mode()[0]
    print(f"Mode: ", mode)
    updated_df_column = df_column.fillna(mode)
    return updated_df_column

for name in mode_impute_col_names:
    df_1[name] = mode_impute(df_1[name])
```

```python
message = "Before Mode Impute:"
divider = "=" * len(message)
print(message + '\n' + divider)
df[mode_impute_col_names].isnull().sum()
```

```python
message = "After Mode Impute:"
print(message, '\n', "="*len(message), sep='')
df_1[mode_impute_col_names].isnull().sum()
```

```python
"""
NaN Replacement
"""
df_2 = df_1.copy()
col_names = ['max_glu_serum', 'A1Cresult']

for name in col_names:
    df_2[name] = df_2[name].replace(np.nan, 'Unknown')
```

```python
message = "Before NaN Replacement:"
divider = "=" * len(message)
print(message + '\n' + divider)
df_1[col_names].isnull().sum()
```

```python
message = "After NaN Replacement:"
divider = "=" * len(message)
print(message + '\n' + divider)
df_2[col_names].isnull().sum()
```
