```python
df = pd.read_csv("training_data.csv")

print(df.shape)
df.head()
```

```python
"""
Replacing the missing data with NaN
"""

missing_cols = ['race', 'weight', 'payer_code', 'medical_specialty', 'diag_1', 'diag_2', 'diag_3', 'gender']

column_to_missing = {}
for col in missing_cols:
    if col != 'gender':
        column_to_missing[col] = '?'
    else:
        column_to_missing[col] = 'Unknown/Invalid'

# print(column_to_missing)

for col, miss_val in column_to_missing.items():
    df[col] = df[col].replace(miss_val, np.nan)

df.isnull().sum()
```

```python
"""
Replace NaN Race entries with mode
"""

# print(df['race'].unique())
mode = df['race'].mode()[0]
df['race'] = df['race'].fillna(mode)
```

```python
"""
Replace NaN Weight entries with median
"""

# print(df['weight'].unique())
sorted_unique_values = ['[0-25)', '[25-50)', '[50-75)', '[75-100)', '[100-125)',
                        '[125-150)', '[150-175)', '[175-200)' '>200']

median = sorted_unique_values[len(sorted_unique_values) // 2]
df['weight'] = df['weight'].fillna(median)
```

```python
"""
Replace NaN payer_code entries with mode
"""

# print(df['payer_code'].unique())
mode = df['payer_code'].mode()[0]
df['payer_code'] = df['payer_code'].fillna(mode)
```

```python
"""
Replace NaN medical_specialty entries with mode
"""

# print(df['medical_specialty'].unique())
mode = df['medical_specialty'].mode()[0]
df['medical_specialty'] = df['medical_specialty'].fillna(mode)
```

```python
"""
Replace NaN max_glu_serum entries with median
"""

# print(df['max_glu_serum'].unique())
sorted_unique_values = ['Norm', '>200', '>300']

median = sorted_unique_values[len(sorted_unique_values) // 2]
df['max_glu_serum'] = df['max_glu_serum'].fillna(median)
```

```python
"""
Replace NaN A1Cresult entries with median
"""

# print(df['A1Cresult'].unique())
sorted_unique_values = ['Norm', '>7', '>8']

median = sorted_unique_values[len(sorted_unique_values) // 2]
df['A1Cresult'] = df['A1Cresult'].fillna(median)
```

```python
"""
Drop rows with missing gender, diag_1, diag_2, and diag_3 as they account
for a very small portion of the data
"""

df.dropna(subset=['gender', 'diag_1', 'diag_2', 'diag_3'], inplace=True)
```
