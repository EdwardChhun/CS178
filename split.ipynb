```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

seed = 1234
np.random.seed(seed)
```

```python
diabetes = pd.read_csv("./diabetes/diabetic_data.csv")
```

```python
data_train, data_test = train_test_split(diabetes, train_size=.9, test_size=.1, shuffle=True, random_state=seed)
```

```python
pd_train = pd.DataFrame(data_train)
pd_test = pd.DataFrame(data_test)

pd_train.to_csv("training_data.csv")
pd_train.to_csv("testing_data.csv")
```
